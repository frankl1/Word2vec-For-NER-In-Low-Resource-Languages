# Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Loading data

```python
results = pd.read_csv("results-corpus-based.csv", index_col=0)
results
```

```python
results_meta = pd.read_csv("results-corpus-based-metadata.csv", index_col=0)
results_meta.head()
```

```python
# we are going to focus only on the projection performances

results_projection = results.loc[["P_ewo", "R_ewo", "F1-ewo"]]
results_projection
```

```python
# transposing the dataframe
res = results_projection.T
res.head()
```

```python
results_meta
```

```python
res.index = res.index.map(int)
results_meta.index = results_meta.index.map(int)
res_good = res.join(results_meta)
res_good
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(res_good.corr(), annot=True)
```

```python
res_good.sort_values(by="h1_size", inplace=True)
res_good.plot(kind="line", y=["F1-ewo", "P_ewo", "R_ewo"], x="h1_size")
```

```python
res_good.sort_values(by="h2_size", inplace=True)
res_good.plot(kind="line", y=["F1-ewo", "P_ewo", "R_ewo"], x="h2_size")
```

```python
res_good.sort_values(by="iteration", inplace=True)
res_good.plot(kind="line", y=["F1-ewo", "P_ewo", "R_ewo"], x="iteration")
```
