# Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Loading data

```python
results = pd.read_csv("results-corpus-based.csv", index_col=0)
results
```

```python
results_meta = pd.read_csv("results-corpus-based-metadata.csv", index_col=0)
results_meta.head()
```

```python
# we are going to focus only on the projection performances

results_projection = results.loc[["P_ewo", "R_ewo", "F1-ewo"]]
results_projection
```

```python
# transposing the dataframe
res = results_projection.T
res.head()
```

```python
results_meta
```

```python
res.index = res.index.map(int)
results_meta.index = results_meta.index.map(int)
res_good = res.join(results_meta)
res_good
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(res_good.corr(method="spearman"), annot=True)
```

```python
res_good.sort_values(by="h1_size", inplace=True)
res_good.plot(kind="line", y=["F1-ewo", "P_ewo", "R_ewo"], x="h1_size")
```

```python
res_good.sort_values(by="h2_size", inplace=True)
res_good.plot(kind="line", y=["F1-ewo", "P_ewo", "R_ewo"], x="h2_size")
```

```python
res_good.sort_values(by="iteration", inplace=True)
res_good.plot(kind="line", y=["F1-ewo", "P_ewo", "R_ewo"], x="iteration")
```

Le rappel de projection semble etre très corrélé avec la taille des couches et le nombre de fold. Le nombre d'itérations n'influence pas énormement, ce qui est normal car les écart-types ont montré que le model est robuste aux initialisation aléatoire. 

Les méilleurs precisions et F1-mesure sont obtenus avec peu de neurones par couche, en l'occurence 350 pour h1 et 250 pour h2, par contre le rappel seulement de 60% traduisant beaucoup de FP

En augmentant le nombre de neurone pas couche, on réduit le taux de FP et on obtient un model assez équilibré avec la précision, le rappel et le F1-mesure supérieurs à 70%
